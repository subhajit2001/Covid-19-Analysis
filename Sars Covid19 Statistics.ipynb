```python
#fig_model=plt.figure(figsize=(18,12))
fig_model=plt.scatter(compare['China Growth Rate per 100,000'],compare['India Growth Rate per 100,000'],marker="x")
X = compare['China Growth Rate per 100,000'].values.reshape(-1,1)
y = compare['India Growth Rate per 100,000'].values.reshape(-1,1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.9, random_state=0)
regressor = LinearRegression()  
regressor.fit(X_train, y_train) #training the algorithm
#To retrieve the intercept:
print(regressor.intercept_)
#For retrieving the slope:
print(regressor.coef_)
Y=regressor.coef_*X+regressor.intercept_
plt.plot(X, Y)
```

<h3><b>Growth Factor of Novel Coronavirus Daily Deaths</b></h3>
<p>Growth factor is the factor by which a quantity multiplies itself over time. The formula used is the <b>change in every day's new deaths (difference in no of new death cases between 2 consecutive days) / new deaths on the previous day</b>. For example, a quantity growing by 7% every period (in this case daily) has a growth factor of 1.07.

A growth factor above 1 indicates an increase, whereas one between 0 and 1 it is a sign of decline, with the quantity eventually becoming zero.

A growth factor below 1 (or above 1 but trending downward) is a positive sign, whereas a growth factor constantly above 1 is the sign of exponential growth.</p>

<p>Maximum of the time the graph has a coordinate above 1 which means that deaths are happening very rapidly with peaks hitting death rates above 2.0 on two occassions which means doubling of the number of new cases everyday. Till now the figure is not so much satisfying and shows a <b>maximum of 2.17</b> and <b>minimum of 0.51</b>.</p>
